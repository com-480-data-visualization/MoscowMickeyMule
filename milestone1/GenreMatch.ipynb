### Merging

```python
merge_test = pd.concat([df_imbd_basics_copy, df_52k_copy, df_lens_movies_copy])
```

```python
merge_test
```

```python

merge_test["title"] = merge_test["title"].str.replace("\"", "")
merge_test["title"] = merge_test["title"].str.replace("\'", "")

merge_test = merge_test.sort_values(ascending= True, by='title')
```

```python
merge_test["runtime"] = merge_test["runtime"].fillna("-1").replace("\\N", "-1")
```

```python
merge_test = merge_test.dropna(subset=["title"])
merge_test["votes"] = merge_test["votes"].fillna(0.0)
```

```python
merge_test["rating"] = merge_test["rating"]*2
merge_test["rating"] = merge_test["rating"].fillna(0)
merge_test["IMBD_rating"] = merge_test["IMBD_rating"].fillna(0)
merge_test["52k_rating"] = merge_test["52k_rating"].fillna(0)
```

```python
merge_test["average_rating"] = merge_test[["IMBD_rating", "52k_rating", "rating"]].mean(axis=1)
merge_test = merge_test.sort_values(ascending= False, by='average_rating')
merge_test.head(25)
```

```python
merge_test["runtime"] = pd.to_numeric(merge_test["runtime"], errors="coerce")
merge_test["IMBD_rating"] = pd.to_numeric(merge_test["IMBD_rating"], errors="coerce")
merge_test["52k_rating"] = pd.to_numeric(merge_test["52k_rating"], errors="coerce")
merge_test["rating"] = pd.to_numeric(merge_test["rating"], errors="coerce")
merge_test["average_rating"] = pd.to_numeric(merge_test["average_rating"], errors="coerce")
```

```python
merge_test
```

```python
combined_entries = merge_test.groupby(["title", "year"], as_index=False).agg({
    "runtime": "max",  # Average runtime
    "genres": lambda x: ', '.join(sorted(set(genre.strip() for genre in ', '.join(x).split(',')))),  # Combine unique genres
    "IMBD_rating": "max",  # Average IMBD rating
    "52k_rating": "max",  # Average 52k rating
    "rating": "max",  # Average rating
    "votes": "sum",  # Total votes
    "production_companies": lambda x: ', '.join(
        sorted(set(company.strip() for company in x if pd.notna(company) and company.lower() != 'nan'))
    ), 
})

# Display the combined DataFrame
combined_entries
```

```python
# Define function named compute_average

def compute_average_rating(row):
    rating_1 = row["IMBD_rating"]
    rating_2 = row["52k_rating"]
    rating_3 = row["rating"]

    count = 0
    total = 0
    if rating_1 > 0:
        total += rating_1
        count += 1
    if rating_2 > 0:
        total += rating_2
        count += 1
    if rating_3 > 0:
        total += rating_3
        count += 1

    if count == 0:
        return 0
    
    return total / count
# Apply the function to the DataFrame
combined_entries["average_rating"] = combined_entries.apply(compute_average_rating, axis=1)
combined_entries = combined_entries.sort_values(ascending= False, by='average_rating')
combined_entries.head(25)
```

```python
combined_entries["genres"] = combined_entries["genres"].str.split(", ")
combined_entries["production_companies"] = combined_entries["production_companies"].str.split(", ")
combined_entries
```

```python
# Step 1: Calculate the mean rating across all movies
C = combined_entries["IMBD_rating"].mean()

# Step 2: Define the minimum number of votes required (e.g., 90th percentile)
m = combined_entries["votes"].quantile(0.95)

# Step 3: Filter movies that qualify (votes >= m)
qualified_movies = combined_entries[combined_entries["votes"] >= m]

# Step 4: Calculate the weighted rating
qualified_movies["weighted_rating"] = (
    (qualified_movies["votes"] / (qualified_movies["votes"] + m) * qualified_movies["IMBD_rating"]) +
    (m / (qualified_movies["votes"] + m) * C)
)

# Step 5: Sort movies by weighted rating
qualified_movies = qualified_movies.sort_values(by="weighted_rating", ascending=False)

# Seperate genres from string to array

qualified_movies.head(10)
```

```python
combined_entries.to_json('genreMatchAll.json', orient='records', indent=2)
qualified_movies.to_json('genreMatchTopRatings.json', orient='records', indent=2)
```
